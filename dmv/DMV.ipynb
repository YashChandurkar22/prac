handle_outliers

```python
def handle_outliers_with_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify and handle outliers
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

    return data, outliers
```

```python
df, outliers = handle_outliers_with_iqr(df, 'sale_price')
outliers
```

Use ARIMA model for forecasting

```python
from statsmodels.tsa.arima.model import ARIMA


order = (1, 1, 1)
model = ARIMA(df['ClosingPrice'], order=order)
model_fit = model.fit()

print(model_fit.summary())


forecast_period = 30
forecast = model_fit.forecast(steps=forecast_period)


plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['ClosingPrice'], label='Observed Price')
plt.plot(pd.date_range(start=df['Date'].iloc[-1], periods=forecast_period+1, closed='right'), forecast, label='Forecasted Price', color='orange')
plt.title('ARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```
